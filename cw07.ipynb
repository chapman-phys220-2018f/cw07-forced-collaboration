## Benchmarking Runtimes

```python
g = c.gen_gaussian_array(20,10,n=1000)
%timeit g = range(100000)
```

```python
g = c.gen_gaussian_list(20,10,n=1000)
%timeit g = range(100000)
```

```python
sc = c.gen_sinc_array(20,10,n=1000)
%timeit sc = range(100000)
```

```python
sc = c.gen_sinc_list(20,10,n=1000)
%timeit sc = range(100000)
```

```python
sf = c.gen_sinf_array(20,10,n=1000)
%timeit sf = range(100000)
```

```python
sf = c.gen_sinf_list(20,10,n=1000)
%timeit sf = range(100000)
```

As we expected, each of the array generators ran faster than the list generators.
